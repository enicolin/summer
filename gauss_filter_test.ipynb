```python
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage.filters import gaussian_filter1d as gaussf1d

#np.random.seed(1756)
x = np.linspace(0,3,20)
y_actual = (x-2)**3 + 1.5*(x-2)**2 + 1
y_noisy = y_actual + np.random.normal(-0.1,0.2,len(x))

sig = 2
x_gauss1d = gaussf1d(x, sig)
y_gauss1d = gaussf1d(y_noisy, sig)

f, ax = plt.subplots(1, figsize = (7,4))
#ax.plot(x,y_actual, color = 'r')
ax.scatter(x,y_noisy)
ax.plot(x_gauss1d, y_gauss1d, color = 'orange')
plt.show()
```
